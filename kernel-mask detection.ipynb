```python
from tensorflow.keras.applications.mobilenet import preprocess_input
train_path = 'Temp/'
train_datagen = ImageDataGenerator(rescale = 1./255, preprocessing_function = preprocess_input, validation_split = 0.25)

train_generator = train_datagen.flow_from_directory(train_path,
                                                  target_size = (224,224),
                                                  color_mode = 'rgb',
                                                  batch_size = 16,
                                                  class_mode = 'categorical',
                                                  subset = 'training',
                                                  shuffle = True)

validation_generator = train_datagen.flow_from_directory(train_path,
                                                 target_size=(224,224),
                                                 color_mode='rgb',
                                                 batch_size=16,
                                                 class_mode='categorical',
                                                 subset = 'validation',
                                                 shuffle=True)

step_size_train=train_generator.n//train_generator.batch_size
step_size_validation=validation_generator.n//validation_generator.batch_size
print(step_size_train)
print(step_size_validation)
```

```python
from tensorflow.keras.applications import MobileNetV2

baseModel = MobileNetV2(weights="imagenet", include_top=False,
    input_tensor=Input(shape=(224, 224, 3)))

headModel = baseModel.output
headModel = AveragePooling2D(pool_size=(7, 7))(headModel)
headModel = Flatten(name="flatten")(headModel)
headModel = Dense(128, activation="relu")(headModel)
headModel = Dropout(0.5)(headModel)
headModel = Dense(8, activation="softmax")(headModel)

model = Model(inputs=baseModel.input, outputs=headModel)
model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
model.summary()
```

```python
class myCallback(tf.keras.callbacks.Callback):
    
    def on_epoch_end(self, epochs, logs={}):
        
        if(logs.get('val_accuracy')>0.99):
            print("\nReached 99% accuracy so training stopped!")
            self.model.stop_training = True
            
callbacks = myCallback()
            
graph = model.fit_generator(generator = train_generator,
                           validation_data = validation_generator,
                           steps_per_epoch = step_size_train,
                           validation_steps = step_size_validation,
                           epochs = 20,
                           callbacks = [callbacks])

model.save("model.h5")
```

```python
def prepare(ima):
    IMG_SIZE = 224
    face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
    gray = cv2.cvtColor(ima, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, 1.1, 4)
    for (x, y, w, h) in faces:
        ima = ima[y:y+h, x:x+w]
    img_array=ima/255.0 
    new_array = cv2.resize(img_array, (IMG_SIZE, IMG_SIZE,3)) 
```
